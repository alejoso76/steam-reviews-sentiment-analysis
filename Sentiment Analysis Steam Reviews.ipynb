## Introducción

### Procesamiento de Lenguaje Natural
El procesamiento de lenguaje natural es un proceso clave para la utilización de datos generados por seres humanos, esto con el fin de hacerlos entendibles para una máquina, y de esta manera, hacerlos útiles en aplicaciones tecnológicas. 
Dicho proceso conlleva una serie de etapas, siendo la primera la tokenización. La tokenización es "el proceso de convertir nuestras secuencias de caracteres, palabras o párrafos en inputs para la computadora." (Murzone, 2020) [1]

Posteriormente viene la etapa de eliminar stop-words, que es simplemente eliminar tokens que sean stop-words. Entiéndase por stop-word palabras sin significado, como lo son los artículos, pronombres, preposiciones. [2] 

En este punto se llega a la etapa de normalizar el texto, aquí se presentan dos opciones: stemming o lematización. Stemming " el proceso de reducir la inflexión de las palabras a sus formas de raíz, como asignar un grupo de palabras a la misma raíz, incluso si la raíz en sí no es una palabra válida en el idioma" (Wikipedia contributors, 2021) [3].
En el caso particular de este proyecto, se optó por la segunda opción, es decir, lematizar. A diferencia del stemming, el proceso de lematización reduce las palabras a una raíz que sí exista en el lenguaje [3]. De esta manera se asegura que todos los tokens serán palabras válidas.

Finalmente, está el proceso de representación vectorial de dichos tokens. Acá, lo que se hace es tener un pool de todos los posibles tokens, cada uno de estos tokens representa una posición dentro de un vector. Luego, se evalúa cada una de las frases o párrafos que ya pasó por las etapas anteriores y se hace su representación en el vector (1 si el token aparece o 0 si no aparece).

### Análisis de Sentimientos
Así pues, el propósito de procesar texto (En este caso), es realizar un análisis de los sentimientos encontrados en dicho texto. El propósito del análisis de sentimientos es realizar una tarea de clasificación masiva de documentos de manera automática, que se centra en clasificar dichos documentos en función de la connotación positiva o negativa del lenguaje dentro de dichos documentos [4].

Con base a este análisis que se desea conducir, se debe elegir un método que permita realizar dicha clasificación, el método elegido es el Machine Learning.

### Machine Learning
Machine Learning (Con siglas ML), es una rama de la inteligencia artificial que permita que las máquinas aprendan por sí solas, sin ser explícitamente programadas para ello [5]. Más que decir que la máquina aprende por sí misma, lo que realmente sucede es que un algoritmo, o una serie de algoritmos, revisan los datos y le dan a la máquina la capacidad de predecir comportamientos futuros. Esto implica que estos sistemas se mejoran de manera autónoma con el tiempo, sin intervención humana alguna [6].

Tomando como base el problema que se desea solucionar y el tipo de clasificación de sentimientos, se eligió el modelo de regresión logística para abordar dicho problema.

### Regresión Logística
La regresión logística es un modelo de Machine Learning que toma su nombre de la función usada en su núcleo, la función logística. Dicha función también es llamada la función sigmoide, inicialmente fue ideada como por estadísticos para describir propiedades del crecimiento de la población en ecología. Es una curva con forma de S y puede tomar cualquier número de valor real y asignarlo a un valor entre 0 y 1, pero nunca esos valores.

La regresión logística, similar a la regresión lineal, usa una ecuación como la representación. Los valores de entrada **x** son combinados linearmente mediante el uso de pesos, para así predecir en valor de salida **y**. La diferencia respecto a la regresión linear es que los valores de salida son modelados usando valores binarios (0-1)  [7]. Es por esto que es el análisis de regresión apropiado para realizar cuando la variable dependiente es dicotómica (Binaria) [8].

Para esto, se descartan las reseñas con caracteres que no hagan parte de los primeros 128 caracteres ASCII. De esta manera se eliminan registros en japonés, chino, español, etc.

```python
pattern = r'[^\x00-\x7F]+'
pattern2 = r'/[\u3000-\u303F]|[\u3040-\u309F]|[\u30A0-\u30FF]|[\uFF00-\uFFEF]|[\u4E00-\u9FAF]|[\u2605-\u2606]|[\u2190-\u2195]|\u203B/'
pattern3 = r'([\u0E00-\u0E7F]+)'
def findNonAscii(df):
    for index, rec in df.iterrows():
        if re.findall(pattern, str(rec['review'])) or re.findall(pattern2, str(rec['review'])) or re.findall(pattern3, str(rec['review'])):
            #df = df.drop([df.index[index]])
            df = df.drop(df[df.id == rec['id']].index)
            df = df.reset_index(drop=True)
            print(F"{rec['id']}   {rec['review']}\n")
            
findNonAscii(steamDataFilter)
```

Función que obtiene le etiqueta que entrega NLTK y la traduce a Wordnet.

```python
def get_wordnet_pos(treebank_tag):

    if treebank_tag.startswith('J'):
        return wordnet.ADJ
    elif treebank_tag.startswith('V'):
        return wordnet.VERB
    elif treebank_tag.startswith('N'):
        return wordnet.NOUN
    elif treebank_tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.VERB
```

```python
def decontracted(phrase):
    # Specific
    #phrase = re.sub(r"won\'t", "will not", phrase)
    #phrase = re.sub(r"can\'t", "can not", phrase)

    # General
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " be", phrase)
    phrase = re.sub(r"\'s", " be", phrase)
    phrase = re.sub(r"\'d", " will", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " be", phrase)
    return phrase
```

```python
stop_words = set(stopwords.words('english'))
def cleanReviews (df):
    reviews_tokens = []
    for review in tqdm(df['review']):
        review_text = review
        #expanded_words = []    
        #for word in review_text.split():
          # using contractions.fix to expand the shotened words
          #expanded_words.append(contractions.fix(word))   

        #review_text = ' '.join(expanded_words)
        
        #pattern = r'[0-9]'
        pattern = r'[?.\-,;."|°¬&¿?=\[\]{}}`_:¡!()^]'

        review_text = re.sub(pattern, '', review_text)
        
        words = word_tokenize(review_text.lower())
        words = [w for w in words if not w in stop_words]
        tagged = nltk.pos_tag(words)
        tokens_lemma = [lemmatizer.lemmatize(token[0], get_wordnet_pos(token[1])) for token in tagged]
        token_index = 0
        for token in tokens_lemma:
            tokens_lemma[token_index] = decontracted(token)
            token_index += 1
        #tokens_fixed = [contractions.fix(token)
        #            for token in words if contractions.fix(token) != 's']
        #lemma_words = [lemmatizer.lemmatize(i) for i in words]
        
        reviews_tokens.append(tokens_lemma)
        
    return reviews_tokens
```

Dato curioso: Se decidió no eliminar números ni "/" porque en el mundo de los videojuegos es común escribir en reseñas valores del tipo: 7/10, esto con el fin de dar una calificación numérica al juego o a algún aspecto del juego en cuestión.

```python
experimentReviews = cleanReviews(experimentData)
```

```python
print(experimentReviews[2])
```

Se regresa la lista tokenizada a una cadena, esto con el fin de representarla en forma de vector posteriormente.

```python
def tokensToStr(token_List):
    reviewsProcessed = []
    for review in token_List:
        r = " ".join(map(str, review))
        reviewsProcessed.append(r)
    return reviewsProcessed
```

```python
reviewsProcessed = tokensToStr(experimentReviews)
```

```python
print(reviewsProcessed[2])
```

```python
y = experimentData['recommendation']
```

```python
X_train, X_test, y_train, y_test = train_test_split(reviewsProcessed, y, random_state=0)
```

### Representacion vectorial

Haciendo uso de la función CountVectorizer() de sklearn, se convierte una colección de documentos de texto, en este caso las reseñas ya tokenizadas y lematizadas, en una matriz de recuentos de tokens. Also similar a la imagen siguiente:

El algoritmo obtiene la lista de listas de tokens de cada reseña y hace una representación vectorial de cada reseña.

```python
cv = CountVectorizer()
ctmTr = cv.fit_transform(X_train)
X_test_dtm = cv.transform(X_test)
```

## Regresión Logística

Se tomó la decisión de elegir el modelo de **Regresión Logística** para realizar las predicciones, esto con base a que dicho modelo es ídoneo cuando la variable dependiente es dicotómica (Binaria). Como lo es el caso de este proyecto: Recomendada o No Recomendada.

De igual manera, se eligió como solver el algoritmo de Broyden–Fletcher–Goldfarb–Shanno, esto a causa de que el problema no tiene restricciones.

```python
model = LogisticRegression(solver='lbfgs',max_iter=10000)
model.fit(ctmTr, y_train)
```

## Predicción

Ya con el modelo entrenado previamente, se procede a hacer la predicción de cada reseña.

```python
y_pred_class = model.predict(X_test_dtm)
```

Se define una función para contar cuántas reseñas fueron negativas y cuántas positivas.

```python
def countPosNegReviews(pred_list):
    positive = 0
    negative = 0
    for element in pred_list:
        if element == 1:
            positive += 1
    return positive, len(pred_list) - positive
```

```python
positiveReviews, negativeReviews = countPosNegReviews(y_pred_class)
print(F'Pred reseñas positivas = {positiveReviews}')
print(F'Pred reseñas negativas = {negativeReviews}')
print(F'Total predicciones = {len(y_pred_class)}')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
labels = ['Recommended', 'Not Recommended']
students = [positiveReviews,negativeReviews]
ax.bar(labels,students,color=['green', 'red'])
plt.show()
```

## Resultados y Discusión

### Matriz de Confusión

La matriz de confusión es una herramienta que permite la visualización del desempeño de un algoritmo de aprendizaje supervisado. Dicha matriz tiene cuatro opciones TP, FP FN y TN.

Las cuatro opciones en esta aplicación son:
1. Reseña positiva y el modelo la clasificó como reseña positiva. Esto sería un verdadero positivo o TP .
2. Reseña negativa y el modelo la clasificó como reseña negativa. Este seria un verdadero negativo o sea un TN.
3. Reseña positiva y el modelo la clasificó como reseña negativa.  Este seria un error tipo II o falso negativo o FN.
4. Reseña negativa y el modelo la clasificó como reseña positiva. Este es un error tipo I,  o falso positivo o FP.


```python
confMatrix = confusion_matrix(y_test, y_pred_class)
```

En este caso se tiene que: 

```python
tn, fp, fn, tp = confMatrix.ravel()
print(F'TP = {tp}')
print(F'FP = {fp}')
print(F'TN = {tn}')
print(F'FN = {fn}')      
```
